```python
import sqlite3
import pandas as pd
```

```python
#Load the data and connect to sqlite db
df= pd.read_csv('../data/cleaned_walmart_sales.csv')
conn= sqlite3.connect('../data/walmart.db')
df.to_sql('Sales', conn, index=False, if_exists='replace')
```

```python
df.head()
```

```python
# Query 1 - Total Overall Sales
query = "select round(sum(Weekly_Sales),2) as Total_Sales from Sales;"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# Query 2 - Average Weekly Sales
query= "select round(avg(Weekly_Sales),2) as Average_Sales from Sales"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
df.head()
```

```python
# Query 3 - Sales by Store
query = "Select Store ,round(sum(Weekly_Sales),2) as Sales from Sales group by Store order by Sales desc limit 10"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# Query 4 - Sales by Department
query = "select Dept, round(sum(Weekly_Sales),2) as Total_Sales from Sales group by Dept order by Total_Sales limit 10"
result = pd.read_sql_query(query, conn)
print(result)
```

```python
# Query 5 - Holiday Sales
query = "select IsHoliday, round(sum(Weekly_Sales),2) as Total_Sales from Sales group by IsHoliday"
result= pd.read_sql_query(query, conn)
print(result)
```

```python
# Query for aggregrated monthly sales for every Store and Dept and saved for further dashboarding.
query = """
SELECT 
    Store, 
    Dept, 
    strftime('%Y-%m', Date) AS YearMonth, 
    SUM(Weekly_Sales) AS Monthly_Sales
FROM sales
GROUP BY Store, Dept, YearMonth;
"""
df_sql = pd.read_sql_query(query, conn)
df_sql.to_csv('../data/monthly_sales_by_store_dept.csv', index=False)
print(df_sql.head())
```
